```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
#Heat map
sns.heatmap(df.corr())
```

```python
#Scatter Plot
sns.FacetGrid(df, hue="Species", palette="husl", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
#Boxplot
sns.boxplot(x="Species", y="PetalLengthCm", palette="husl", data=df)
```

```python
#Pair plot
sns.pairplot(df.drop("Id", axis=1), hue="Species", palette="husl", size=3)
```

```python
#count plot
sns.countplot(df['Species'])
plt.show()  
```

```python
df.drop("Id", axis=1).boxplot(by="Species", figsize=(10, 10))
plt.show()
```
